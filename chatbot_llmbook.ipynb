```python
import os
from dotenv import load_dotenv, find_dotenv

_ = load_dotenv(find_dotenv())

from langchain.chains import LLMChain, ConversationChain
from langchain.chat_models import ChatOpenAI
from langchain.prompts import ChatPromptTemplate

llm = ChatOpenAI(temperature = 0.9)

prompt = ChatPromptTemplate.from_template(
    """You are a helpful assisant that help the user to plan an optimized itinerary. 
    Here is the itinerary, Suggest the optimal itinerary 
    {itinerary}"""

)

chain = LLMChain(llm=llm, prompt=prompt)
itinerary = "I'm going to Rome for 2 days, what can I visit"
#result = chain.run(itinerary)
#print(result)

#Add Memory
from langchain.memory import ConversationBufferMemory
from langchain.chains.conversation.base import ConversationChain

memory = ConversationBufferMemory()

conversation = ConversationChain(
    llm = llm,
    verbose = True,
    memory = memory
)
print(conversation.run("Hi There"))
print(conversation.run("What is the most iconic place in Rome"))
```

```python
conversation.run("What kind of other events")
```

```python
while True:
    query = input('you: ')
    if query == 'q':
        break
    output = conversation({"input": query})
    print('User: ', query)
    print('AI: ', output['response'])
```

```python
from langchain.llms import OpenAI
from langchain.chat_models import ChatOpenAI
from langchain.embeddings.openai import OpenAIEmbeddings
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.vectorstores import FAISS
from langchain.document_loaders import PyPDFLoader
from langchain.chains import ConversationalRetrievalChain
from langchain.memory import ConversationBufferMemory

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size = 1500,
    chunk_overlap = 200
)

raw_documents = PyPDFLoader('italy_travel.pdf').load()
documents = text_splitter.split_documents(raw_documents)
db = FAISS.from_documents(documents, OpenAIEmbeddings())
memory = ConversationBufferMemory(
    memory_key = 'chat_history',
    return_messages=True
)

llm = ChatOpenAI()
```

```python
qa_chain = ConversationalRetrievalChain.from_llm(llm, retriever=db.as_retriever(), memory=memory, verbose=True)
qa_chain.run({'question':'Give me some review about the pantheon'})
```

```python
from langchain.agents.agent_toolkits import create_retriever_tool

tool = create_retriever_tool(
    db.as_retriever(),
    "italy_travel",
    "Searches and returns documents regarding Italy"
)

tools = [tool]

memory = ConversationBufferMemory(memory_key='chat_history', return_messages=True)

from langchain.agents.agent_toolkits import create_conversational_retrieval_agent

from langchain.chat_models import ChatOpenAI
llm = ChatOpenAI(temperature=0)

agent_executor = create_conversational_retrieval_agent(llm, tools, memory_key='chat_history', verbose=True)
```

```python
agent_executor({"input": "Tell me something about Pantheon"})
```

```python
output = agent_executor({"input": "what can I visit in India in 3 days"})
```

```python
from langchain import SerpAPIWrapper
search = SerpAPIWrapper()

tools = [
    tool.from_function(
        func = search.run,
        name = "Search",
        description="Useful for when you need to answer questions about current events"
    ),
    create_retriever_tool(
        db.as_retriever(),
        "italy_travel",
        "Searches and return documens regarding Italy"
    )
]
agent_executor = create_conversational_retrieval_agent(llm, tools, memory_key='chat_history', verbose=False)
```

```python
output1 = agent_executor({"input": "who are playing the finals of Copa America 2024"})
print(output1['output'])
output2 = agent_executor({"input": "What can I visit in India in 3 days"})
print(output2['output'])
output3 = agent_executor({"input": "I'm traveling to Italy, can you give me some suggestions for the main attractions to visit"})
print(output3['output'])
```
